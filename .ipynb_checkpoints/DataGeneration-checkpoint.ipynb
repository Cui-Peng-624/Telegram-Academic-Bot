# 生成数据

```python
import re

# 定义函数以生成和写入数据
def generate_and_write_data(prompt, label, filename):
    completion = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": prompt}
        ]
    )
    
    generated_text = completion.choices[0].message.content.strip()
    
    # 分割生成的请求和标签
    lines = re.sub(r'^\d+\.\s*', '', generated_text, flags=re.MULTILINE).split('\n') # 去掉前面的1，2，3
    data = [(line.split(' -> ')[0].strip(), line.split(' -> ')[1].strip()) for line in lines if '->' in line]
    
    # 写入CSV文件
    with open(filename, 'a', newline='', encoding='utf-8') as csvfile:
        csvwriter = csv.writer(csvfile)
        for text, label in data:
            csvwriter.writerow([text, label])
```

```python
import csv

# 初始化CSV文件
with open('data/training_data.csv', 'w', newline='', encoding='utf-8') as csvfile:
    csvwriter = csv.writer(csvfile)
    csvwriter.writerow(['text', 'label'])
```

```python
from tqdm import tqdm

# 分领域生成数据并写入CSV
epochs = 30

for _ in tqdm(range(epochs)): # 生成5*10*epochs行的数据
    generate_and_write_data(education_prompt, 'Education', 'data/training_data.csv')
    generate_and_write_data(computer_science_prompt, 'Computer Science', 'data/training_data.csv')
    generate_and_write_data(medical_prompt, 'Medicine', 'data/training_data.csv')
    generate_and_write_data(literature_prompt, 'Literature', 'data/training_data.csv')
    generate_and_write_data(other_prompt, 'Other', 'data/training_data.csv')
```
